```python
import sqlite3
import re
import nltk
import os

nltk.download('punkt')

# Ask for the database file at the start
db_file = input("📂 Enter the path to your invoice database (e.g., invoice.db): ")

# Check if the file exists
if not os.path.exists(db_file):
    print(f"❌ Error: The file '{db_file}' does not exist. Please check and restart.")
    exit()

# Function to fetch invoice details from the external database
def get_invoice_details(invoice_number):
    conn = sqlite3.connect(db_file)  # Connect to the user-provided database
    cursor = conn.cursor()

    try:
        cursor.execute("SELECT * FROM Invoices WHERE invoice_id = ?", (invoice_number,))
        result = cursor.fetchone()
    except sqlite3.Error as e:
        return f"⚠️ Database error: {e}"

    conn.close()

    if result:
        return f"📜 Invoice #{result[0]}\n🗓 Date: {result[3]}\n💰 Amount: ${result[2]}"
    else:
        return "⚠️ Invoice not found. Please check the number."

# Function to handle customer queries
def chatbot_response(user_input):
    user_input = user_input.lower()

    # Check for greetings
    if any(word in user_input for word in ["hello", "hi", "hey"]):
        return "🤖 Hello! How can I assist you today?"

    # Check for invoice request
    invoice_match = re.search(r"invoice\s*#?\s*(\d+)", user_input)
    if invoice_match:
        invoice_number = invoice_match.group(1)
        return get_invoice_details(invoice_number)

    # Default response
    return "🤖 Sorry, I didn't understand. You can ask me about invoices by saying 'Check my invoice #123'."

# Start chatbot loop
print("🤖 Customer Support Chatbot (type 'exit' to stop)")
while True:
    user_query = input("You: ")
    if user_query.lower() == "exit":
        print("🤖 Goodbye!")
        break
    response = chatbot_response(user_query)
    print("Bot:", response)
```
