```python
import sqlite3
import re
import nltk

nltk.download('punkt')

# Function to fetch invoice details from the database
def get_invoice_details(invoice_number):
    conn = sqlite3.connect("invoices.db")
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM invoices WHERE invoice_number = ?", (invoice_number,))
    result = cursor.fetchone()

    conn.close()

    if result:
        return f"üìú Invoice #{result[1]}\nüóì Date: {result[2]}\nüí∞ Amount: ${result[3]}"
    else:
        return "‚ö†Ô∏è Invoice not found. Please check the number."

# Function to handle customer queries
def chatbot_response(user_input):
    user_input = user_input.lower()

    # Check for greetings
    if any(word in user_input for word in ["hello", "hi", "hey"]):
        return "ü§ñ Hello! How can I assist you today?"

    # Check for invoice request
    invoice_match = re.search(r"invoice\s*#?\s*(\d+)", user_input)
    if invoice_match:
        invoice_number = invoice_match.group(1)
        return get_invoice_details(invoice_number)

    # Default response
    return "ü§ñ Sorry, I didn't understand. You can ask me about invoices by saying 'Check my invoice #123'."

# Start chatbot loop
print("ü§ñ Customer Support Chatbot (type 'exit' to stop)")
while True:
    user_query = input("You: ")
    if user_query.lower() == "exit":
        print("ü§ñ Goodbye!")
        break
    response = chatbot_response(user_query)
    print("Bot:", response)
```

```python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect("invoices.db")
cursor = conn.cursor()

# Create the invoices table
cursor.execute("""
CREATE TABLE IF NOT EXISTS invoices (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    invoice_number TEXT UNIQUE,
    date TEXT,
    amount REAL
)
""")

# Insert sample invoices (change these if needed)
sample_data = [
    ("101", "2024-02-20", 150.00),
    ("102", "2024-02-21", 200.50),
    ("103", "2024-02-22", 99.99)
]

for data in sample_data:
    cursor.execute("INSERT OR IGNORE INTO invoices (invoice_number, date, amount) VALUES (?, ?, ?)", data)

conn.commit()
conn.close()

print("‚úÖ Database and sample invoices created successfully!")
```
